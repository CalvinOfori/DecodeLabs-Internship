```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import requests
import io

import re
from urllib.parse import urlparse, parse_qs, unquote

%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/mikeasilva/data-scientist-skills/master/raw_skills.csv"
r = requests.get(url)

try:
  df = pd.read_csv(io.StringIO(r.text))
  items = df['raw_skills'].dropna().astype(str).tolist()
except Exception as e:
  print(f"Error reading CSV: {e}")

df.head(20)
```

```python
invalid_values = ["['']", "['n/a']", "[]"]
df = df[~df['raw_skills'].isin(invalid_values)]
df.dropna(subset=['raw_skills'], inplace=True)
df.drop('id', axis=1, inplace=True)
df.head(10)
```

```python
def extract_job_title(url_string):
  url_string = str(url_string)
  url_string = unquote(url_string)

  parsed_url =  urlparse(url_string)
  query_params = parse_qs(parsed_url.query)

  for key in ['q', 'query', 'job', 'title']:
    if key in query_params:
      if key in query_params:
            clean_text = query_params[key][0]
            return clean_text.title()

    last_part = parsed_url.path.strip('/').split('/')[-1]

    clean_text = re.sub(r'[^a-zA-Z]', ' ', last_part)

    clean_text = ' '.join(clean_text.split()).title()

    return clean_text

df['job_title'] = df['url'].apply(extract_job_title)

# 4. Now you can safely drop the messy URL column if you want
df.drop(columns=["url"], inplace=True)

# Display the first 5 rows to verify it worked!
df.head(10)
```

```python
df.dropna(subset=['raw_skills'], inplace=True)
df.head(10)
```

```python
df.dropna(subset=['job_title'], inplace=True)

df = df[df['job_title'].str.len() >= 6]

df = df.reset_index(drop=True)

print(f"Dataset loaded and cleaned! Total valid rows: {len(df)}")
df.head(10)
```

```python
print("="*10)
print("Provide at least 3 technical skills")
user_input = input("Enter your skills (comma-separated): ")
```

```python
if not user_input.strip():
  print("\n[COLD START DETECTED]: No user profile data found.")
  print("Bypassing similarity math.")
  print("Recommended Career Paths:")

  trending = df['raw_skills'].value_counts().head(3)
  print(trending)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer()

item_vectors = vectorizer.fit_transform(df['raw_skills'])
user_vector = vectorizer.transform([user_input])

similarity_scores = cosine_similarity(user_vector, item_vectors)
print("Similarity Scores:", similarity_scores)
```

```python
job_titles = ["Data Scientist", "DevOps Engineer", "Backend Developer"]

scores = similarity_scores.flatten()

sorted_indices = scores.argsort()[::-1]
top_n = 3

top_n_indices = sorted_indices[:top_n]

print(f"--- Top {top_n} Recommended Career Paths ---")
for index in top_n_indices:
    career_path = df['job_title'].iloc[index]
    matched_skills = df['raw_skills'].iloc[index]
    score = scores[index]

    print(f"Match Score: {score * 100:.1f}%")
    print(f"Recommended Role: {career_path}")
    print(f"Required Skills: {matched_skills}\n")
```
