# **SIMPLE CHATBOT - DECODELABS TASK 1**

```python
responses = {
    'hello': "Hi! I'm Aura. Your Personal virtual assistant",
    'bye': 'Goodbye! Have a great day.',
    'help': 'I can assist you with basic queries. What do you need?',
    'status': 'All systems are currently operational.',
    'project': 'You are currently working on Project 1: Rule-Based AI Chatbot.',
    'joke': 'Why do programmers prefer dark mode? Because light attracts bugs!'

}

print("Initiating Logic Engine. Type 'exit' to quit")
```

```python
while True:
  raw_input = input("You: ")
  clean_input = raw_input.lower().strip()

  if clean_input == 'exit':
    print("Thank you for your time. Goodbye")
    break

  reply = responses.get(clean_input, "I do not understand your input. Please Repeat yourself")

  print(f"Aura: {reply}")
```
